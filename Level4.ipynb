# Question 1

List the top 10 tracks in the global throughout year 2017 with their total stream counts.

```python
def getyear(arr):
    return arr.split('-')[0]
arr = "2017-01-01"
data['year'] = data['Date'].map(getyear)
```

```python
data.groupby(['year','Track Name']).agg({'Streams': sum}).sort_values('Streams',ascending = False).head(10)
```

# Question 2

List the top 10 artists (or groups) those has the most stream counts for all their tracks combined with the stream counts of each of their tracks.

```python
topartist = data.groupby('Artist')\
                .agg({'Streams': sum})\
                .sort_values('Streams',ascending = False)\
                .head(10)\
                .reset_index()\
                .rename(columns = {'Streams': 'Streamssum'})
```

```python
topartist
```

# Question 3

List the top 10 tracks in December, 2017 for each continent (North America, Europe, Asia, South America, Oceania).

```python
data2017 = data[data['year']=='2017']
def getmonth(arr):
    return arr.split('-')[1]
data2017['Month'] = data2017['Date'].map(getmonth)
data2017[data2017['Month'] == '12']
```

```python
country = pd.read_json('countries.json')
country
```

```python
def up(regionstr):
    regionstr =regionstr.upper()
    return regionstr

#     return country[regionstr]

data2017['Region'] = data2017['Region'].map(up)
```

```python
# country[data2017['Region'][0]]
country = country.loc['continent'].reset_index()
```

```python
country.columns = ['Region','continent']
```

```python
datawithcontinent = pd.merge(data2017, country)
level43 = datawithcontinent[datawithcontinent['Month']=='12'].groupby(['continent','Track Name'])\
                                                    .agg({'Streams': sum})\
                                                    .reset_index()
```

```python
level43.groupby(['continent']).apply(lambda x: x.nlargest(10,'Streams'))
```
